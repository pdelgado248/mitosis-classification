```python
#Notebook to create .csv files from the .tif videos, where the name of the video and the label are stored.
#It writes also the images as .jpg files to be able to use them in the classifier.

#dataPath = '/content/drive/MyDrive/Pablo Delgado/Nantes-Madrid/ALL DATA/Mitosis classification/Final_images/mitosis_12_train'
dataPath = '/content/drive/MyDrive/Pablo Delgado/Nantes-Madrid/ALL DATA/Mitosis classification/Final_images/mitosis_12_test'

resultsPath = '/content/drive/MyDrive/Pablo Delgado/Nantes-Madrid/Publications/Mitosis classifier/Publications code/Rebuttal'


import tifffile as tif
from glob import glob as glob
import os
import csv
from PIL import Image


csvFileName = resultsPath + '/testData.csv'

videosResultFolder = resultsPath + '/test-jpg-videos'

header = ['name', 'label']

with open(csvFileName, 'w', newline="") as file:
  csvwriter = csv.writer(file)
  csvwriter.writerow(header)

  for classCount,classFolder in enumerate(glob(dataPath+'/*')):

    if 'Abnormal' in os.path.basename(classFolder):
      classID = 'Abnormal'
    elif 'Normal' in os.path.basename(classFolder):
      classID = 'Normal'

    for videoFullPath in glob(classFolder+'/*'):

      videoName = os.path.basename(videoFullPath)

      print([videoName[:-4],classID])

      csvwriter.writerow([videoName[:-4],classID])

      video = tif.imread(videoFullPath)

      #Create a new folder to store inside it the separate frames of the video
      #videoFolderPath = os.path.join(videosResultFolder,videoName)
      #os.mkdir(videoFolderPath[:-4])

      for frameNum in range(video.shape[0]):
        frame = video[frameNum,:]
      
        imFrame = Image.fromarray(frame)
        finalName = videoFolderPath[:-4]+'/'+ videoName[:-4]+'-'+str(frameNum)+'.jpg'
        imFrame.save( finalName)
        print(videoName[:-4],': Frame ',frameNum,' saved')
```
